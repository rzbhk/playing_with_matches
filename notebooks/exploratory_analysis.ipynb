```python
def plot_trip_duration_distribution(oddf,_bins=50):
    oddf['duration'] = oddf['time_end'] - oddf['time_begin']
    s, loc , scale = st.lognorm.fit(oddf["duration"],fscale=oddf['duration'].mean())

    ax3 = sns.distplot(oddf['duration'],bins=_bins,norm_hist=False,kde=False)
    ticks = ax3.get_yticks()
    plt.close()
    fig , ax = plt.subplots(figsize=(10,6))
    
    lnspace = np.linspace(0,3600,3600)
    ax.plot(lnspace,st.lognorm.pdf(lnspace,s,loc,scale),label="lognorm fit")
    sns.distplot(oddf['duration'],bins=_bins,kde=False,ax=ax,norm_hist=True,fit=st.gamma,fit_kws={"label" : 'gamma fit'})
    ax.set_ylabel("probability",size=20)
    ax.set_xlabel("duration",size=20)
    ax.legend(prop={'size': 20})
    ax2 = ax.twinx()
    ax2.set_yticks(ticks[:-1])
    
```

```python
def plot_trip_distance_distribution(oddf,_bins=50):
    oddf['distance'] = np.sqrt( (oddf.y_begin - oddf.y_end) ** 2 + (oddf.x_begin - oddf.x_end) ** 2 )
    
#     print(oddf.distance.describe())
    s, loc , scale = st.lognorm.fit(oddf["distance"],loc=oddf['distance'].mean(),scale=oddf['distance'].std())

    ax3 = sns.distplot(oddf['distance'],bins=_bins,norm_hist=False,kde=False)
    ticks = ax3.get_yticks()
    plt.close()
    fig , ax = plt.subplots(figsize=(10,6))
    
    
#     rvs = st.lognorm.rvs(s, loc,scale, size=len(oddf), random_state=None)
#     print(rvs)
#     rvs = [ i if i >= 0 and  i <= oddf['distance'].max() else st.lognorm.rvs(s, loc,scale, size=1, random_state=None) for i in rvs]
    
    lnspace = np.linspace(0,35000,3500)
    ax.plot(lnspace,st.lognorm.pdf(lnspace,s,loc,scale),label="lognorm fit")
    sns.distplot(oddf['distance'],bins=_bins,kde=False,ax=ax,norm_hist=True,fit=st.gamma,fit_kws={"label" : 'gamma fit'})
    ax.set_ylabel("probability",size=20)
    ax.set_xlabel("distance",size=20)
    ax.legend(prop={'size': 20})
    ax2 = ax.twinx()
    ax2.set_yticks(ticks[:-1])
    ax2.set_ylabel("count")
```

```python
def cdf_plot(ser,ax=None,figsize=None, label=None):
    ser = ser.sort_values()
    cum_dist = np.linspace(0.,1.,len(ser))
    ser_cdf = pd.Series(cum_dist, index=ser)
    return ser_cdf.plot(drawstyle='steps',figsize=figsize,yticks=np.arange(0.,1.001,0.1),ax=ax, label=label)
```
